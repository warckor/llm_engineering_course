```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from openai import OpenAI
import tiktoken

import os
from dotenv import load_dotenv
load_dotenv()

OPENAI_API_KEY = os.getenv("GITHUB_TOKEN")
OPENAI_BASE_URL = os.getenv("GITHUB_MODEL_URL")
# MODEL = "openai/gpt-4.1-mini"
MODEL = "cohere/cohere-command-a"
# EMBEDDING_MODEL = "openai/text-embedding-3-small"
EMBEDDING_MODEL = "cohere/Cohere-embed-v3-multilingual"
```

```python
client = OpenAI(
    api_key=OPENAI_API_KEY,
    base_url=OPENAI_BASE_URL
)
```

```python
loader = PyPDFLoader("./docs/Resolución exámenes TLP.pdf")
pages = []

async for page in loader.alazy_load():
    pages.append(page)
```

```python
print(f"{pages[0].metadata}\n")
print(pages[2].page_content[:500])
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
chunks = splitter.split_documents(pages)
```

```python
print(f"Número de páginas: {len(pages)}")
print(f"Número de chunks: {len(chunks)}")
print()
print(chunks[19].page_content)
```

```python
content = [chunk.page_content for chunk in chunks]

for con in content:
    if "Haskell" in con:
        print(con)
        print("-"*40)
```

```python
system_prompt = f"""
Eres un asistente que ayuda a responder preguntas sobre un conjunto de documentos.
Eres preciso y conciso en tus respuestas.
Utiliza únicamente la información proporcionada en los documentos.
Si la respuesta no aparece en los documentos, di "Lo siento, en la documentacion no sale nada", no incluyas información adicional.
Siempre responde en el mismo idioma en el que se te pregunta.
Documentos: {content[:36]}

# Ejemplos:
Pregunta: ¿Como se arregla una lavadora?
Respuesta: Lo siento, en la documentacion no sale nada.

Pregunta: ¿Como se prepara una lasaña?
Respuesta: Lo siento, en la documentacion no sale nada.
"""

user_prompt = "salen preguntas sobre algo relacionado con resolucion perezosa o algo asi?"

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
]
```

```python
encoding = tiktoken.get_encoding("cl100k_base")
num_tokens = len(encoding.encode("\n".join(content[:40])))
system_prompt_tokens = len(encoding.encode(system_prompt))
print(f"Número de tokens: {num_tokens}")
print(f"Número de tokens en el system prompt: {system_prompt_tokens}")
```

```python
response = client.chat.completions.create(
    model=MODEL,
    messages=messages,
    max_tokens=500,
    temperature=0.0
)
```

```python
print(response.choices[0].message.content)
```
